# 1. plt.figure()

Creates a new empty figure. 

Use figsize to control output dimensions, important for papers or dashboards 

dpi = 150 (gives sharper screen)

dpi = 300 ( for print)

```python
fig = plt.figure(
    figsize= (10,6),   #width, height in inches 
    dpi = 120,         #dots per inches (screen : 100-120)
    facecolor ='white'
)
plt.show()
```

# 2. plt.subplots(rows, cols)


The go-to for multi panel figures.
it helps to create miltiple charts in 1 figure 


```python
fig, axes = plt.subplots(
    2, 3, #WE MADE A GRID OF 2 ROWS X 3COLUMNS = 6PLOTS 
    figsize = (15, 8), #HERE WIDTH IS 15 AND HEIGHT IS 8 
    sharex = True) # ALL PLOTS SHARE THE SAME X- AXIS SCALE 
# Bottom row shows x-values
# Top row doesn’t repeat them (cleaner view)

#ACCESS INDIVIDUAL AXES :
ax = axes[0, 1] 
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))  # 1 row, 2 columns

axes[0].plot([1,2,3], [4,5,6])
axes[0].set_title('Plot 1')

axes[1].plot([1,2,3], [6,5,4])
axes[1].set_title('Plot 2')

plt.tight_layout()
plt.show()
```

# 3.LINE PLOT 
### plt.plot(x,y) 
The standard visualising training. 

```python
months = [1, 2, 3, 4, 5, 6]
sales = [200, 350, 280, 500, 450, 600] 

plt.plot(months, sales, color = "steelblue", linewidth = 2, marker = "o")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales")
plt.show()
```

# 4. SCATTER PLOT 
### plt.scatter()
plots individual dots - good for seeing relationshiops between two things 


```python
hours_studied = [1, 2, 3, 4, 5, 6, 7, 8] 
exam_score = [40, 50, 55, 65, 70, 80, 85,  90]
plt.scatter(hours_studied, exam_score, color = "coral", s= 80)
plt.xlabel("Hours studied")
plt.ylabel("Exam score")
plt.title("Study hours vs score")
plt.show()
```

# 5.HISTOGRAM 

### plt.hist()

# 6.BAR GRAPH
### VERTICAL VERSION 
### plt.bar()

```python
subjects = ["Math", "Science", "English", "history"] 
score = [85, 92, 78, 88]
plt.bar(subjects, score) 
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.title("Subject scores")
plt.show()
```

### HORIZONTAL VERSION 
### plt.barh()


```python
plt.barh(subjects, score)
plt.title("Subject scores")
plt.xlabel("scores")
plt.ylabel("Subjects")
plt.show()
```

# 7.BOXPLOT
### plt.boxplot()
Shows the spread of data - minimum, maximum, median and outliers 


box plot summarizes your data using 5 key values:

Minimum value 


Q1 (25th percentile)


Median (middle value)


Q3 (75th percentile)


Maximum


```python
math_scores = [55, 60, 70, 75, 80, 85, 90]
science_scores = [65, 70, 72, 78, 82, 88, 95]
english_scores = [50, 55, 60, 65, 68, 72, 75]

plt.boxplot([math_scores, science_scores, english_scores], 
            labels =["Math", "Science", "English"])


plt.ylabel("Scores")
plt.title("Score distribution by subjects")
plt.show()
```

# 8.DISPLAYS GRID OF NUMBERS

# 9.LABELLED LINE GRAPH WITH KEY
### Legend from label= arguments

```python
days = [1, 2, 3, 4, 5]
apples= [10, 12, 9, 14, 11]
bananas = [7, 8, 10, 6, 9]

plt.plot(days, apples, label = "Apples", color = "Red")
plt.plot(days, bananas, label = "Banana", color = "yellow")
plt.legend()
plt.xlabel("Day")
plt.ylabel("Fruits Sold")
plt.title("Daily fruit sales")
plt.show()
```

# 10. ADDING ARROWS USING 
## plt.annotate()

```python
x = list(range(1, 11))
y = [2.5, 1.8, 1.3, 0.9, 0.7, 0.65, 0.63, 0.62, 0.64, 0.70]

plt.plot(x, y, marker = 'o')

plt.annotate('Best model',
             xy=(8, 0.62),           # arrow tip (data point)
             xytext=(5, 0.55),        # text position
             arrowprops=dict(arrowstyle='->', color='red'),
             fontsize=11, color='red')

plt.xlabel('Epoch')
plt.ylabel('Val Loss')
plt.title('Annotating Best Epoch')
plt.show()
```

# 11. ZOOMING INTO SPECIFIC REGIONS 

```python
plt.plot(range(100), np.exp(-np.linspace(0, 5, 100)))
plt.xlim(20, 80)    # show only epochs 20-80
plt.ylim(0, 0.5)    # zoom into low loss region
plt.title('Zoomed Loss Curve')
plt.show()
```

# 13. custom tick labels 

```python
scores = [0.91, 0.87, 0.83, 0.79]
models = ['XGBoost', 'Random Forest', 'Logistic Reg', 'SVM']

plt.bar(models, scores, color='steelblue')
plt.xticks(rotation=30, ha='right')   # rotate labels so they don't overlap
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.tight_layout()
plt.show()
```
